```python
import os
import zipfile
import pandas as pd

os.chdir('../2. Prepared Data')

ES_statistic_df = pd.read_csv('AirBase_ES_v8_statistics.zip', sep='\t', compression='zip')
FR_statistic_df = pd.read_csv('AirBase_FR_v8_statistics.zip', sep='\t', compression='zip')
PT_statistic_df = pd.read_csv('AirBase_PT_v8_statistics.zip', sep='\t', compression='zip')

statistic_df_raw = pd.concat([ES_statistic_df, FR_statistic_df, PT_statistic_df]) #fusion des df

statistic_df = statistic_df_raw[statistic_df_raw.statistic_name == 'annual mean'] #filtre sur la moyenne annuelle
statistic_df = statistic_df.iloc[:, [0,1,2,3,4,7,8,11]] #selection des colonnes utiles

statistic_df.to_csv("statistic_df.csv")
```
